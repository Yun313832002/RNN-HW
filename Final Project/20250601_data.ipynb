```python
import json
import csv

def json_to_csv_dictwriter(json_file_path, csv_file_path, features):
    """
    將 JSON 檔案轉換為 CSV 檔案，使用 DictWriter。

    Args:
        json_file_path (str): 輸入的 JSON 檔案路徑。
        csv_file_path (str): 輸出的 CSV 檔案路徑。
        features (list): CSV 檔案的標題列表。
    """
    try:
        with open(json_file_path, 'r', encoding='utf-8') as json_file:
            data = json.load(json_file) # 如果是單一 JSON 物件，可以直接讀取
            
            # 如果 JSON 檔案是多個 JSON 物件組成的列表，則需要調整
            # 例如：data = [json.loads(line) for line in json_file]

        with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
            writer = csv.DictWriter(csv_file, fieldnames=features)
            writer.writeheader()  # 寫入標題行
            
            # 如果 data 是一個列表，則遍歷列表
            if isinstance(data, list):
                for row in data:
                    writer.writerow(row)
            else: # 如果 data 是一個單一字典
                writer.writerow(data)
        
        print(f"成功將 '{json_file_path}' 轉換為 '{csv_file_path}'")

    except FileNotFoundError:
        print(f"錯誤：找不到檔案 '{json_file_path}'。")
    except json.JSONDecodeError:
        print(f"錯誤：無法解析 JSON 檔案 '{json_file_path}'。請檢查 JSON 格式。")
    except Exception as e:
        print(f"發生未知錯誤：{e}")





# 將範例資料寫入一個 JSON 檔案
with open('train_dataset_list.json', 'w', encoding='utf-8') as f:
    json.dump(train_dataset_list, f, indent=4) # 假設是列表

json_to_csv_dictwriter('train_dataset_list.json', 'train_dataset_list_output.csv', feature_names_list)
```

```python
def process_csv_columns(input_csv_path, output_csv_path):
    """
    從輸入的 CSV 檔案讀取 'ID' 和 'Description' 欄位，
    並將它們寫入新的 CSV 檔案，標題改為 'Name' 和 'Text'。

    Args:
        input_csv_path (str): 原始 CSV 檔案的路徑。
        output_csv_path (str): 目標 CSV 檔案的路徑。
    """
    try:
        with open(input_csv_path, 'r', newline='', encoding='utf-8') as infile:
            reader = csv.reader(infile)
            header = next(reader)  # 讀取標題行

            # 找出 'ID' 和 'Description' 欄位的索引
            try:
                id_index = header.index('ID')
                description_index = header.index('Description')
            except ValueError as e:
                print(f"錯誤：原始 CSV 檔案中缺少必要的欄位。{e}")
                return

            with open(output_csv_path, 'w', newline='', encoding='utf-8') as outfile:
                writer = csv.writer(outfile)

                # 寫入新的標題
                writer.writerow(['Name', 'Text'])

                # 逐行讀取並寫入新檔案
                for row in reader:
                    # 確保行足夠長以包含所需的索引
                    if len(row) > max(id_index, description_index):
                        name_value = row[id_index]
                        text_value = row[description_index]
                        writer.writerow([name_value, text_value])
                    else:
                        print(f"警告：跳過一行，因為它的欄位不足：{row}")

        print(f"成功處理 '{input_csv_path}' 並將結果寫入 '{output_csv_path}'。")

    except FileNotFoundError:
        print(f"錯誤：找不到檔案 '{input_csv_path}'。請檢查路徑是否正確。")
    except Exception as e:
        print(f"處理 CSV 檔案時發生錯誤：{e}")

input_file = 'train_dataset_list_output.csv'
output_file = 'train_dataset_list_output_X.csv'
process_csv_columns(input_file, output_file)
```

```python
import csv

def exclude_columns_from_csv(input_csv_path, output_csv_path, columns_to_exclude):
    """
    從輸入的 CSV 檔案中排除指定的欄位，並將其餘欄位寫入新的 CSV 檔案。

    Args:
        input_csv_path (str): 原始 CSV 檔案的路徑。
        output_csv_path (str): 目標 CSV 檔案的路徑。
        columns_to_exclude (list): 包含要排除的欄位名稱的列表。
    """
    try:
        with open(input_csv_path, 'r', newline='', encoding='utf-8') as infile:
            reader = csv.reader(infile)
            original_header = next(reader)  # 讀取原始標題行

            # 找出要排除的欄位的索引
            exclude_indices = []
            for col_name in columns_to_exclude:
                try:
                    exclude_indices.append(original_header.index(col_name))
                except ValueError:
                    print(f"警告：原始 CSV 檔案中未找到要排除的欄位 '{col_name}'。")
            
            # 建立新的標題 (排除指定欄位)
            new_header = [col for i, col in enumerate(original_header) if i not in exclude_indices]

            with open(output_csv_path, 'w', newline='', encoding='utf-8') as outfile:
                writer = csv.writer(outfile)
                
                # 寫入新的標題行
                writer.writerow(new_header)

                # 逐行讀取並寫入新檔案，排除指定欄位的值
                for row in reader:
                    new_row = [value for i, value in enumerate(row) if i not in exclude_indices]
                    writer.writerow(new_row)

        print(f"成功處理 '{input_csv_path}' 並將結果寫入 '{output_csv_path}'，已排除欄位：{columns_to_exclude}")

    except FileNotFoundError:
        print(f"錯誤：找不到檔案 '{input_csv_path}'。請檢查路徑是否正確。")
    except Exception as e:
        print(f"處理 CSV 檔案時發生錯誤：{e}")


input_file = 'train_dataset_list_output.csv'
output_file = 'train_dataset_list_output_Y.csv'
columns_to_exclude = ['ID', 'Description'] # 你要排除的欄位名稱

exclude_columns_from_csv(input_file, output_file, columns_to_exclude)
```

```python
import pandas as pd
import os

def merge_csv_files(file1_path, file2_path, file3_path, output_file_path):
    """
    讀取三個 CSV 檔案，並將它們的所有欄位按順序合併到一個新的 CSV 檔案中。

    Args:
        file1_path (str): 第一個 CSV 檔案的路徑 (例如：train_dataset_list_output_X.csv)。
        file2_path (str): 第二個 CSV 檔案的路徑 (例如：output_tactics_split.csv)。
        file3_path (str): 第三個 CSV 檔案的路徑 (例如：output_techniques_split.csv)。
        output_file_path (str): 合併後的新 CSV 檔案的輸出路徑。
    """
    try:
        print(f"正在讀取檔案：{file1_path}")
        df1 = pd.read_csv(file1_path)
        print(f"檔案 '{file1_path}' 讀取完成。形狀: {df1.shape}")

        print(f"正在讀取檔案：{file2_path}")
        df2 = pd.read_csv(file2_path)
        print(f"檔案 '{file2_path}' 讀取完成。形狀: {df2.shape}")

        print(f"正在讀取檔案：{file3_path}")
        df3 = pd.read_csv(file3_path)
        print(f"檔案 '{file3_path}' 讀取完成。形狀: {df3.shape}")

        # 檢查所有 DataFrame 的行數是否一致
        if not (len(df1) == len(df2) == len(df3)):
            print("警告: 檔案的行數不一致。這可能導致合併結果不如預期。")
            print(f"檔案1行數: {len(df1)}, 檔案2行數: {len(df2)}, 檔案3行數: {len(df3)}")

        # 使用 pd.concat 沿著列軸 (axis=1) 合併 DataFrame
        # 這會將 df2 和 df3 的所有欄位追加到 df1 的後面
        print("正在合併檔案...")
        merged_df = pd.concat([df1, df2, df3], axis=1)
        print(f"檔案合併完成。新形狀: {merged_df.shape}")

        # 寫入新的 CSV 檔案
        merged_df.to_csv(output_file_path, index=False)
        print(f"所有欄位已成功合併到 '{output_file_path}'。")

    except FileNotFoundError as e:
        print(f"錯誤: 找不到檔案 - {e.filename}。請檢查檔案路徑是否正確。")
    except pd.errors.EmptyDataError:
        print("錯誤: 讀取到空的 CSV 檔案。")
    except Exception as e:
        print(f"合併 CSV 檔案時發生錯誤: {e}")

if __name__ == "__main__":
    # 定義你的檔案路徑
    file1 = 'train_dataset_list_output_X.csv'
    file2 = 'output_tactics_split.csv'
    file3 = 'output_techniques_split.csv'
    output_merged_file = 'merged_all_columns.csv'

    # 為了讓程式碼可運行，先創建一些模擬的 CSV 檔案
    # 實際使用時，請確保這些檔案已存在
    if not os.path.exists(file1):
        pd.DataFrame({'colA': [1,2,3], 'colB': ['x','y','z']}).to_csv(file1, index=False)
        print(f"創建了模擬檔案: {file1}")
    if not os.path.exists(file2):
        pd.DataFrame({'Tactic1': [0,1,0], 'Tactic2': [1,0,1]}).to_csv(file2, index=False)
        print(f"創建了模擬檔案: {file2}")
    if not os.path.exists(file3):
        pd.DataFrame({'TechA': [1,0,1], 'TechB': [0,1,0]}).to_csv(file3, index=False)
        print(f"創建了模擬檔案: {file3}")

    # 呼叫函數進行合併
    merge_csv_files(file1, file2, file3, output_merged_file)
```
